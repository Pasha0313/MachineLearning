```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Part 2 - Data Preprocessing

### Step 1 : Load and Inspect the Dataset

```python
data = pd.read_csv('challenge_dataset.csv')
```

```python
display(data.head())
print('\n')
display(data.tail())
print('\n')
display(data.describe())
print('\n')
display(data.info())
print('\n')
display(data.shape)
print('\n')
display(data.columns)
```

```python
numeric_cols = ['Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'NumberOfDeviceRegistered', 'SatisfactionScore', 
                'NumberOfAddress','OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder',
                 'CashbackAmount'] 
```

### Step 2 : Check Duplicates

```python
duplicates = data.duplicated()
print(f"Number of duplicates: {duplicates.sum()}")

data = data.drop_duplicates()
```

### Step 3 : remove the "CustomerID" colmun

```python
data   = data.iloc[:, 1:]
```

### Step 4 : Check Missing Values

```python
missing_counts = data.isnull().sum()
missing_percentage = (missing_counts / len(data)) * 100
missing_summary = pd.DataFrame({'Missing_Count': missing_counts, 'Missing_Percentage': missing_percentage})
print(missing_summary.sort_values(by='Missing_Percentage', ascending=False))
```

### Step 5-6 : Data Cleaning 

1. Identifying Missing Values
2. method 1 : remove all missing data row
3. unbalance data # Count complete rows for Churn = 0,1
4. Identifying Outliers

```python
data_T = data.copy()
data_T.dropna(inplace=True)
data_T.info()

Q1 = data_T[numeric_cols].quantile(0.25)
Q3 = data_T[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = (data_T[numeric_cols] < (Q1 - 1.5 * IQR)) | (data_T[numeric_cols] > (Q3 + 1.5 * IQR))

data_T_RO = data_T[~outliers.any(axis=1)]

print(f"Original data shape: {data_T.shape}")
print(f"Filtered data shape: {data_T_RO.shape}")
print(f"Number of rows removed: {data_T.shape[0] - data_T_RO.shape[0]}")
```

```python
print(data['Churn'].value_counts())

complete_churn_0 = data[data['Churn'] == 0.0].dropna().shape[0]
complete_churn_1 = data[data['Churn'] == 1.0].dropna().shape[0]

print(f"Complete rows for Churn = 0: {complete_churn_0}")
print(f"Complete rows for Churn = 1: {complete_churn_1}")

Multiple = 2
churn_1_data = data[data['Churn'] == 1.0].dropna()
churn_0_data = data[data['Churn'] == 0.0].dropna().sample(n=complete_churn_1*Multiple, random_state=2024)

data_total = pd.concat([churn_1_data, churn_0_data])
data_total = data_total.sample(frac=1, random_state=2024).reset_index(drop=True)

print(data_total['Churn'].value_counts())

Q1 = data_total[numeric_cols].quantile(0.25)
Q3 = data_total[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = (data_total[numeric_cols] < (Q1 - 1.5 * IQR)) | (data_total[numeric_cols] > (Q3 + 1.5 * IQR))

data_RO = data_total[~outliers.any(axis=1)]

print("Outliers based on IQR:\n", outliers)

data = data_total.copy()

print("\n")
print(f"Original data shape: {data.shape}")
print(f"Filtered data shape: {data_RO.shape}")
print(f"Number of rows removed: {data.shape[0] - data_RO.shape[0]}")
```

##### Remove Outliers

```python
data = data_RO.copy()
data_T = data_T_RO.copy()
```

```python
display(data.head())
print('\n')
display(data.tail())
print('\n')
display(data.describe())
print('\n')
display(data.info())
print('\n')
display(data.shape)
print('\n')
display(data.columns)
```

## Part 3 : Data preparation

Encoding categorical data

One-Hot Encoding the "PreferredLoginDevice" + "PreferredPaymentMode" + "PreferedOrderCat" columns

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), ['PreferredLoginDevice', 'PreferredPaymentMode', 'PreferedOrderCat'])  
    ],
    remainder='passthrough'  
)

data_encoded = np.array(ct.fit_transform(data))
data_encoded_T = np.array(ct.fit_transform(data_T))

encoded_columns = ct.get_feature_names_out(input_features=data.columns)
encoded_columns_T = ct.get_feature_names_out(input_features=data_T.columns)

encoded_columns = [col.split('__')[-1] for col in encoded_columns]  
encoded_columns_T = [col.split('__')[-1] for col in encoded_columns_T]  

data_encoded = pd.DataFrame(data_encoded, columns=encoded_columns)
data_encoded_T = pd.DataFrame(data_encoded_T, columns=encoded_columns_T)
```

```python
display(data_encoded.head())
print('\n')
display(data_encoded.tail())
print('\n')
display(data_encoded.describe())
print('\n')
display(data_encoded.info())
print('\n')
display(data_encoded.shape)
print('\n')
display(data_encoded.columns)
```

## Part 4 - Churn Prediction with Random Forest

## Step 1 : Splitting the dataset into the Training set and Test set

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.model_selection import train_test_split
X = data_encoded.iloc[:, :-1].values
y = data_encoded.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)

X = data_encoded_T.iloc[:, :-1].values
y = data_encoded_T.iloc[:, -1].values
X_train_T, X_test_T, y_train_T, y_test_T = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

Count how many zeros and ones are in y_train and y_test

```python

train_counts = pd.Series(y_train).value_counts()
test_counts = pd.Series(y_test).value_counts()

print("y_train counts:\n", train_counts)
print("\ny_test counts:\n", test_counts)
```

## Step 2 : Initializing the Random Forest

```python
rf = RandomForestClassifier(random_state=2024)  
```

### Step 3 : Training the Random Forest
Step3-1
Find best_params

```python
from sklearn.model_selection import GridSearchCV

param_grid = { 
    'n_estimators': [100, 200, 500],
    'max_features': ['sqrt', 'log2', None],  
    'max_depth': [4, 6, 8],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

CV_rfc = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, verbose=1, n_jobs=-1)
CV_rfc.fit(X_train, y_train)

display("Best Parameters:", CV_rfc.best_params_)
```

Step3-2
Fill with the best_params

```python
rf_final=RandomForestClassifier(
    bootstrap=True,
    criterion='gini',
    max_depth=8, 
    max_features='log2',
    min_samples_leaf=1,
    min_samples_split = 2,
    n_estimators= 500, 
    random_state=2024)

rf_final.fit(X_train, y_train)
```

Make predictions
